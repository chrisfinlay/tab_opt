```python
X = 27.3
Ns = np.arange(3, 2001)
xs = [np.arange(0, 30, 30/N) for N in Ns]
dxs = Ns/30
ys = [np.sin(2.*np.pi*x/X + 0.8) for x in xs]
ys_avg = np.array([y.mean() for y in ys])
```

```python
for x, y in zip(xs, ys):
    plt.plot(x, y, '.')
```
