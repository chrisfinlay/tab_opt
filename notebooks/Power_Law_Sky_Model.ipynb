```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def power_law(I, I_min=1e-4, alpha=1.6):
    
    return (I/I_min)**-alpha
```

```python
I = np.logspace(-4, 0, 1001)
```

```python
plt.loglog(I, power_law(I))
```

```python
def pdf(I, I0=1e-4, alpha=1.6):
    
    factor = (alpha - 1.)
    return factor * I0**factor * I**-alpha
```

```python
plt.loglog(I, pdf(I))
```

```python
def cdf(I, I0=1e-4, alpha=1.6):
    
    return 1 - (I/I0)**(1.-alpha)
```

```python
plt.semilogx(I, cdf(I))
```

```python
def icdf(rand_unif, I0=1e-4, alpha=1.6):
    
    return I0 * (1. - rand_unif)**( 1. / (1. - alpha) )
```

```python
P = np.random.random(1_000)
```

```python
I_ = icdf(P)
```

```python
plt.hist(I_, bins=10_000, density=True);
plt.loglog(I, pdf(I))
```
